# Optimal countries for living

This notebook will compute the most optimal country for living conditions by my standards using the CIA wordfact book dataset.
Each country will get a score that will be affected by the following columns. Population density; The biggest area per person will be the most optimal, Birth rate; The smallest birth rate will be the most optimal, Real GDP per capita; the highest will be optimal. Geographic conditions; I will choose hot countries that are less than 35 degrees latitude.


```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv(r"C:\Users\farfu\notebooks\datasets\countries.csv")

#vartriable that will activate a courdinates function
cord = False

# with GDP
#df = df[["Country","Geography: Area - total", "People and Society: Population", "People and Society: Birth rate", "Economy: Real GDP per capita"]]

# without GDP
# df = df[["Country","Geography: Area - total", "People and Society: Population", "People and Society: Birth rate"]]

# with cordinats less the 35 for nice climete and GDP
df = df[["Country","Geography: Area - total", "People and Society: Population", "People and Society: Birth rate", "Economy: Real GDP per capita","Geography: Geographic coordinates"]]
cord = True
```

## cleaning the data

```python
def extract_first_numeric(cell):
    # Check if the input is a string
    if isinstance(cell, str):
        # Use a regular expression to find the first numeric value
        match = re.search(r'([\d,.]+)', cell)
        
        # If a match is found and the group is not empty, convert to numeric; otherwise, return NaN
        if match and match.group():
            try:
                return float(match.group().replace(',', ''))
            except ValueError:
                return np.nan
        else:
            return np.nan
    else:
        return np.nan
    
# Loop through all columns and convert values to numeric
for column in df.columns:
    print(column)
    if column != "Country":
        # Apply the extraction function to the specified column
        df[column] = df[column].apply(extract_first_numeric)
        
        # loop cells and drop ones that have courdinets that are bigger then 35N and 35S
        if cord == True:
            if column == "Geography: Geographic coordinates":
                for row in range(df.shape[0]):
                    if df.loc[row,column] > 35:                       
                        df = df.drop(row)
                        
        
        

#         df[column] = pd.to_numeric(df[column].str.extract(r'([\d.]+)').iloc[:, 0], errors='coerce')
#         df[column] = df[column].str.extract(r'([\d.]+)')
#         df[column] = pd.to_numeric(df[column], errors='coerce')
```

```python
# create population densety column
df["Population densety"] = df["Geography: Area - total"] / df["People and Society: Population"]  
```

## Normalizing the data by rank or by max scaled

```python
# copy the data 
df_n = df.copy() 
  
# apply normalization techniques 

# normaliztion with max scaled techniqe (dident give a prortional result)
# for column in df_n.columns:
#     if column != "Country":
#         df_n[column] = df_n[column]  / df_n[column].abs().max() 


# Normalize each column based on its order

# Function to normalize a column based on its order rank
def normalize_by_order(column):
    ranked_values = column.rank(na_option='bottom', method='min')
    normalized_values = ranked_values / len(column)
    return normalized_values

# Normalize each column based on its order rank
for column in df_n.columns:
    if column != "Country":
        df_n[column] = normalize_by_order(df_n[column])
        
        
```

```python
# drop col that is not relevent
df_n = df_n.drop(["People and Society: Population", "Geography: Area - total"],axis="columns")
```

```python
# convert birth rate column to fit lower birth rate = larger number
df_n["People and Society: Birth rate"] = 1 - df_n["People and Society: Birth rate"]
```

## coucalate the best countreis 

```python
# # make a general score for each country

# Create an empty "Score" column
df_n["Score"] = 0

for column in df_n.columns:
    if column != "Country":
        df_n["Score"] += df_n[column]

```

```python
# Display the top 10 scored countries
top_countries = df_n.nlargest(40,"Score")
print(top_countries)
```

```python
top_indices = top_countries.index.tolist()
rearranged_df = df.reindex(top_indices)
rearranged_df.head
```
